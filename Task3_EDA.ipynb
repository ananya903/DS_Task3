###  *Exploratory Data Analysis

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['Sales'],color='black',label='Sales',shade=True)
sns.kdeplot(df['Profit'],color='red',label='Profit',shade=True)
plt.xlim([-100,1000])
plt.legend()
```

Profit is more than sales but in some cases its vice-versa. 

```python
sns.lineplot(x='Discount',y='Profit',data = df)
plt.show()
```

Profit decreases with increase in discount

```python
df.corr()
```

```python
sns.heatmap(df.corr(),cmap='rocket_r',annot=True,linecolor='white',linewidth=2,cbar=False)
```

Sales and Profit are Moderately Correlated.

Discount and Profit are Negatively Correlated.

Quantity and sales are also correlated but less than moderate. 

```python
sns.pairplot(df,hue='Category',palette='husl')
```

```python
sns.pairplot(df,hue='Region')
```

```python
#Count  plot of each column 
import warnings
warnings.filterwarnings("ignore")

fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,7));

sns.countplot(df['Category'],ax=axs[0][0])
sns.countplot(df['Segment'],ax=axs[0][1])
sns.countplot(df['Ship Mode'],ax=axs[1][0])
sns.countplot(df['Region'],ax=axs[1][1])
axs[0][0].set_title('Category',fontsize=20)
axs[0][1].set_title('Segment',fontsize=20)
axs[1][0].set_title('Ship Mode',fontsize=20)
axs[1][1].set_title('Region',fontsize=20)
plt.tight_layout()

```

```python
plt.figure(figsize=(20,10))
sns.countplot(df['Sub-Category'])
plt.title('Sub-Category',fontsize=20)
```

```python
plt.figure(figsize=(18,10))
sns.countplot(df['State'])
plt.xticks(rotation=90)
plt.title('State',fontsize=20)
```

```python
plt.figure(figsize=(18,10))
sns.countplot(df['Discount'])
plt.xticks(rotation=90)
plt.title('Discount',fontsize=20)
```

```python
#Distribution of the data using the plot
```

```python
fig, axs = plt.subplots(ncols=2, nrows = 2, figsize = (20,10))
sns.distplot(df['Sales'], color = 'red',  ax = axs[0][0])
sns.distplot(df['Profit'], color = 'cyan',  ax = axs[0][1])
sns.distplot(df['Quantity'], color = 'magenta',  ax = axs[1][0])
sns.distplot(df['Discount'], color = 'blue',  ax = axs[1][1])
axs[0][0].set_title('Sales Distribution', fontsize = 20)
axs[0][1].set_title('Profit Distribution', fontsize = 20)
axs[1][0].set_title('Quantity distribution', fontsize = 20)
axs[1][1].set_title('Discount Distribution', fontsize = 20)
plt.show()
```

```python
dfa = df['State'].value_counts()
dfa.head(20)
```

```python
dfa.plot(kind='pie',figsize=(35,25),autopct = '%1.1f%%')
plt.ylabel('Frequency / Number of deals',fontsize=20)
plt.xlabel('States',fontsize=20)

plt.title('State Wise Dealings', fontsize = 20)
plt.show()
```

```python
df['State'].value_counts().mean()
```

```python
dfb = df['City'].value_counts()
dfb=dfb.head(50)
dfb.plot(kind='pie',figsize=(25,25),autopct = '%1.1f%%')
plt.ylabel('Frequency / Number of deals',fontsize=20)
plt.xlabel('City',fontsize=20)

plt.title('City Wise Dealings', fontsize = 20)
plt.show()
```

```python
df['City'].value_counts().mean()
```

```python
Sales = df.groupby('Region').sum()['Sales'].sort_values(ascending = False)
Profits = df.groupby('Region').sum()['Profit'].sort_values(ascending = False)
plt.style.use('seaborn')
Sales.plot(kind='bar', figsize=(10,5),fontsize = 15, color = 'Red')
Profits.plot(kind='bar', figsize=(10,5),fontsize = 15 , color='Black')
plt.xlabel('Regions',fontsize =  20)
plt.ylabel('Total sales & profit',fontsize = 20)
plt.title("Sales vs Profit",fontsize = 20)
plt.show()
```

```python
#Citywise Analysis of the Profit

df_city= df.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean()
df_city = df_city.sort_values('Profit')
df_city.head()
```

```python
df_city['Profit'].head(30).plot(kind='bar',figsize=(15,5),color = 'cyan')
plt.title('City wise analysis of Sale, Discount, profit')
```

```python
df_city['Profit'].tail(30).plot(kind='bar',figsize=(15,5),color = 'cyan')
plt.title('City wise analysis of Sale, Discount, profit')
```

```python
df_state= df.groupby([ 'State' ])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(10)
```

```python
df_state1=df_state.sort_values('Profit')

df_state1[['Profit']].plot(kind = 'bar', figsize = (15,4))
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
plt.show()
```

```python
df_category = df.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()
df_category
```

```python
df_category.plot.pie(subplots=True, 
                     figsize=(18, 20), 
                     autopct='%1.1f%%', 
                     labels = df_category.index)
```

```python
p = df.groupby('Sub-Category').min()['Profit'].sort_values()
```

```python
plt.figure(figsize=(8,6))
p.plot.bar()
plt.title('Sub-Categories with highest loss')
plt.xticks(rotation = 60 , fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

```python
#Sub-Category wise Sales, Profit and Discount

df_sub_cat = df.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
df_sub_cat.head(10)
```

```python
plt.figure(figsize = (10,10))
plt.pie(df_sub_cat['Sales'], labels = df_sub_cat.index, autopct = '%1.1f%%')
plt.title('Sales Analysis(Sub-Category Wise)', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.pie(df_sub_cat['Discount'], labels = df_sub_cat.index, autopct = '%1.1f%%')
plt.title('Discount Analysis(Sub-Category Wise)', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```
